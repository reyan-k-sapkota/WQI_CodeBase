```python


count = 0


Area_name = 'Sacramento'


def details(area):
    count = 0
    for i in range(num_rows_colArea):
        if df['Area'][i] == area:
            count +=1
            
    area_a_data = df[df['Area'] == area]
    mean_value = area_a_data['CCME_Values'].mean()
    std_val = area_a_data['CCME_Values'].std()
    median_val = area_a_data['CCME_Values'].median()
    
    return count, mean_value, std_val, median_val

print(f"Number of times {Area_name} appears in 'Area' column: {details(Area_name)[0]}")
print(f"Mean value for {Area_name}:", details(Area_name)[1])
print(f"SD value for {Area_name}:", details(Area_name)[2])
print(f"Median value for {Area_name}:", details(Area_name)[3])

```
